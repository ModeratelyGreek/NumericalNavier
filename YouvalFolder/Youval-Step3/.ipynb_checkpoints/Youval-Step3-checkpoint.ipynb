```julia
using Pkg
Pkg.add("Plots")
using Plots
```

```julia
nx = 41  # number of grid poins
dx = 2/(nx-1)  # if our spatial domain is (0,2)
nt = 20  # number of timesteps we want to calculate
nu = 0.3 # value of viscosity
sigma = 0.2 # sigma is a parameter we learn more about later
dt = sigma*dx^2/nu  # amount of time each tmiestep covers

# setting initial conditions
u = ones(nx)
u[Int(floor(0.5 / dx)) + 1:Int(floor(1 / dx)) + 1] .= 2
print(u)

x = range(0, stop=2, length=nx)
plot(x, u, legend=false)
```

```julia
un = ones(nx)  # initialize temporary array

for n in 1:nt  # Loop over each timestep
    un = copy(u)  # Copy the current state of u into un
    for i in 2:nx-1  # Iterate over spatial grid points, starting from 2 because Julia is 1-indexed
        u[i] = un[i] + nu * dt / dx^2 * (un[i+1] - 2 * un[i] + un[i-1])
    end
end

x = range(0, stop=2, length=nx)
plot(x, u, legend=false)
```

```julia
function animate_wave_propagation_adaptive_steps(nx=200, max_nt=400, sigma=0.3, nu=0.2)
    dx = 2.0 / (nx-1) # Spatial step size
    dt = sigma*dx^2/nu
    u = ones(nx) # Initial condition, wave at rest

    # Set initial wave pulse
    u[Int(floor(0.5 / dx)) + 1 : Int(floor(1 / dx)) + 1] .= 2

    anim = @animate for nt in 1:max_nt
        max_u = maximum(u)
        dt = min(dx / max_u, dt) # Adaptive time step based on CFL condition
        Courant = max_u * dt / dx # Calculate Courant number
        #@show dt
        @show Courant

        un = copy(u) # Copy the current state of u into un

        for i in 2:nx-1 # Iterate over spatial grid points, leaving out the first point for boundary condition
            # Finite difference method for convection equation
            u[i] = un[i] + nu * dt / dx^2 * (un[i+1] - 2 * un[i] + un[i-1])
        end

        x = range(0, stop=1, length=nx) # Spatial domain
        plot(x, u, legend=false, ylims=(0, 3)) # Plot current state with fixed y-axis limits
    end

    gif(anim, "wave_propagation_animation_adaptive_steps.gif", fps=10) # Save the animation as a GIF file
end
```
